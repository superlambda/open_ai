```python
from openai import OpenAI
import os

api_key = 'sk-ttt-XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX'
client = OpenAI(api_key=api_key)
COMPLETION_MODEL = "gpt-3.5-turbo-instruct"

def ask_gpt3(prompt):
    response = client.completions.create(
        model=COMPLETION_MODEL,
        prompt=prompt,
        max_tokens=512,
        n=1,
        stop=None,
        temperature=0.5,
    )
    message = response.choices[0].text.strip()
    return message
print("你好，我是一个聊天机器人，请你提出你的问题吧?")
questions = []
answers = []
def generate_prompt(prompt, questions, answers):
    num = len(answers)
    for i in range(num):
        prompt += "\n Q : " + questions[i]
        prompt += "\n A : " + answers[i]
    prompt += "\n Q : " + questions[num] + "\n A : "        
    return prompt
while True:
    user_input = input("> ")
    questions.append(user_input)
    if user_input.lower() in ["bye", "goodbye", "exit"]:
        print("Goodbye!")
        break
    
    prompt = generate_prompt("", questions, answers)
    answer = ask_gpt3(prompt)
    print(answer)
    answers.append(answer)
```
